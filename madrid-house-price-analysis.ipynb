```python
# Import libariry
import pandas as pd 
import numpy as np 
from matplotlib import pyplot as plt 
import seaborn as sns   

# Configure pandas display options for better readability
pd.options.display.float_format = '{:,.1f}'.format
pd.options.display.max_columns = None
pd.options.display.max_rows = None

# Set seaborn style for visualizations
#sns.set()
```

```python
# Load the Madrid housing dataset
data = pd.read_csv('E:\DATA ANALYSIS - Cources\L- Projects Profile\Python\madrid houses price analysis\houses_Madrid.csv')
data.info()
```

# **Data Preprocessing**

```python
# Remove columns with all NaN values
data = data.dropna(axis= 1, how= 'all')
data.info()
```

```python
# check For duplicate rows
data.duplicated().sum()
```

```python
data.columns
```

```python
# taking only needed columns
df = data[['id', 'subtitle', 'sq_mt_built', 'sq_mt_useful', 'n_rooms', 'n_bathrooms',
            'n_floors', 'sq_mt_allotment', 'rent_price', 'buy_price', 'buy_price_by_area','built_year', 'parking_price']]
df.info()
```

```python
# Rename 'subtitle' column to 'city'
df = df.rename(columns= {'subtitle':'city' })
```

```python
# Keep only essential columns with fewer missing values  
df = df[['id', 'city', 'sq_mt_built', 'n_rooms', 'n_bathrooms', 'rent_price', 'buy_price', 'buy_price_by_area']]
df.info()
```

```python
# Display summary statistics for the number of rooms
df.n_rooms.describe()
```

```python
# Remove rows where number of rooms is zero
df.drop(df.index[df.n_rooms == 0], axis= 0, inplace= True)
```

```python
# Calculate mode of bathrooms column
df.n_bathrooms.mode()
```

```python
# Fill NaN values in bathrooms column with mode (1) and replace 0 with 1
df.n_bathrooms.fillna(1, inplace= True)
df.n_bathrooms = df.n_bathrooms.replace(0,1)
```

```python
df.info()
```

```python
# Convert n_bathrooms to integer type 
df.n_bathrooms = df.n_bathrooms.astype(int)
```

```python
# Remove rows with missing sq_mt_built values
df.dropna(subset= 'sq_mt_built', inplace= True)
```

```python
df.info()
```

```python
df.sample(10)
```

```python
# Replace negative values in rent_price with zero 
for value in df.rent_price.values:
    if value < 0:
        df.rent_price = 0
```

```python
# split column city to (district, city)
df['district'] = df.city.str.split(',', expand= True)[0]
```

```python
df.head(10)
```

```python
df.columns
```

```python
# Reorder columns for clarity
df = df[['id', 'district', 'sq_mt_built', 'n_rooms', 'n_bathrooms', 'rent_price',
       'buy_price', 'buy_price_by_area']]
df.head()
```

```python
data.head()
```

```python
# house_type_id is_new_development
data.is_new_development.unique()
```

```python
# Remove rent_price column as it's not used in further analysis
df.drop(columns= ['rent_price'], inplace= True)
```

# **Modeling**

## **linear regression model**

```python
# Filter data for Recoletos district
reco = df.query('district == "Recoletos"')
```

```python
reco.sample(10)
```

```python
# Fit linear regression model for Recoletos district
import statsmodels.api as sm

reco['intercept'] = 1
linear_model = sm.OLS(reco['buy_price'], reco[['intercept', 'sq_mt_built']])
result = linear_model.fit()
result.summary()
```

```python
# Predict price for a 300 m² flat in Recoletos
# yhat = b0 + b1 * x

x= 300
price = 43420 + 8975 * x
print("predicted price of flat in recoletos with are of {} m2 = {} euro".format(x, price))
```
